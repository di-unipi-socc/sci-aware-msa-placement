```python
from pathlib import Path
import pandas as pd

# Load the data
base_path = Path("../parquets")

df_path = base_path / "raw-sci-aware.parquet"
df = pd.read_parquet(df_path)

# count how many cols starts with placement/
placement_cols = [col for col in df.columns if col.startswith("placement")]
# delete all cols that starts with placement/
config_cols = [col for col in df.columns if col.startswith("config/experiment/")]
# rename "base" to "exhaustive" in the heuristic column
df["heuristic"] = df["heuristic"].replace({"base": "exhaustive"})
df = df.drop(columns=placement_cols)
df = df.drop(columns=config_cols)
df.set_index("trial_id", inplace=True)
df.to_parquet(base_path / "clean-sci-aware.parquet")
```
